## 一个关于全球空军力量分析的例子

```r
MAdata=read.csv("TotalAircraftStrengthData2015.csv")
MAdata
```

```r
printf <- function(...) invisible(print(sprintf(...)))
printf("2015年统计 %d 个国家的空军力量",length(MAdata$AircraftStrength))
```

### 2015年空军力量最强的国家

```r
MAdata$Country[which.max(MAdata$AircraftStrength)]
```

```r
printf("军力均值%f，中位数%f",mean (MAdata$AircraftStrength),median (MAdata$AircraftStrength))
printf("军力四分位数%f",quantile (MAdata$AircraftStrength))
```

```r
boxplot(MAdata$AircraftStrength)
```

```r
printf("中国空军2015军力%f",MAdata$AircraftStrength[which(grepl("China", MAdata$Country))])
```

```r
MAdataO<-MAdata[order(MAdata$AircraftStrength,decreasing=T),]
```

### 2015各国空军军力排序

```r
MAdataO
```

```r
MAdataO$AircraftStrength[1]>sum(MAdataO$AircraftStrength[2:7])
```

## 2015各国空军军力分析结果
* 被统计的126个国家中，中国空军军力排名第三为2942，是所有国家空军军力均值的7倍以上。
* 然而排名第一的美国空军军力为13444，超过中国的4倍以上，比排名第二到第7的总合还高。
* 世界空军军力分布处于极端不平衡状态，有一半国家的空军军力低于125。
